**The FinanceToolkit** is an open-source toolkit in which all relevant financial ratios (100+), indicators and performance measurements are written down in the most simplistic way allowing for complete transparency of the calculation method. This allows you to not have to rely on metrics from other providers and, given a financial statement, allow for efficient manual calculations. This leads to one uniform method of calculation being applied that is available and understood by everyone.

````python
from financetoolkit import Portfolio

API_KEY = "FINANCIAL_MODELING_PREP_API_KEY"
````

The Portfolio Module is meant to be used for the analysis of your own portfolio. This can be a portfolio of stocks, bonds, real estate or any other type of asset. The module allows you to create overviews to understand performance (as opposed to a benchmark) and see your profit-and-loss over time. It also allows you to directly connect to the Toolkit functionalities itself meaning that you can calculate any of the 150+ metrics for the assets in your portfolio and your actual portfolio (weighted average over time).

**It is important to note that you will need a specific XLSX, XLS or CSV template in which you provide the data of your portfolio. This template can be acquired by not providing a `portfolio_dataset` parameter as follows.**

````python
portfolio = Portfolio(api_key=API_KEY, benchmark_ticker="^GSPC")
````

As written above, it provides an example portfolio and places two files, `portfolio_template.xlsx` and `portfolio_template.csv`, in the current working directory. This template can be used to fill in your own portfolio data. The template consists of the following columns:

- `Date`: The date of the transaction (e.g. 2021-01-01)
- `Identifier`: The ticker of the traded asset (e.g. AAPL)
- `Currency`: The currency of the asset (e.g. USD)
- `Volume`: The amount of shares or contracts (e.g. 10 for buy orders, and -10 for sell orders)
- `Price`: The price of the asset at the time of the transaction (e.g. 100)
- `Costs`: The costs of the transaction (e.g. -3)

For this Notebook, we proceed with the example portfolio that is provided in the template. We will calculate the metrics for this portfolio and provide an overview of the performance. Let's start off by getting an overal impression of the assets in the portfolio and their performance.

````python
portfolio.get_portfolio_overview(include_portfolio=True)
````

The above overview demonstrates how each of your assets has performed and is an aggregation of all your transactions. The same transactions have occurred against the benchmark, which is the S&P 500 in this case. The benchmark is used to compare the performance of your portfolio against a broad market index which gives an indication whether your decisions have been fruitful. At the bottom you can find the `Portfolio` value which shows the performance of your portfolio itself, weighted by the assets in the portfolio.

The same overview can also be obtained for each transaction with `get_transactions_overview` which also demonstrates the Profit and Loss (PnL) of your transactions. This can be useful to see which transactions have been profitable and which have not. Here, you have the option to define the PnL method which can be either FIFO (First In First Out), LIFO (Last In First Out) or Average (Average Cost). By default, FIFO is used.

````python
portfolio.get_transactions_overview(pnl_method="FIFO")
````

For both the portfolio and the transactions overviews, you can also see the performance per period. This can be `yearly`, `quarterly`, `monthly`, `weekly` or `daily`. This gives you deeper insights on the performance of your portfolio and your transactions in a specified period.

````python
display(portfolio.get_portfolio_performance(period="yearly"))

display(portfolio.get_transactions_performance(period="quarterly"))
````

Ultimately, what makes this module powerful is the ability to directly connect to the Toolkit functionalities. This means that you can calculate any of the 150+ metrics for the assets in your portfolio and your actual portfolio (weighted average over time). This can be done as follows where it is important you assign the toolkit to a new variable to prevent constant API calls.

````python
toolkit = portfolio.toolkit
````

Then, it is possible to calculate anything but let's start with acquiring the Net Profit Margin ratio for the assets in the portfolio. It is important to note that the example portfolio contains ETFs as well meaning not for all tickers the ratios can be calculated.

````python
net_profit_margin = toolkit.ratios.get_net_profit_margin()

display(net_profit_margin)

net_profit_margin.loc["Portfolio"].plot(
    title="Net Profit Margin of the Portfolio",
    kind="bar",
    figsize=(15, 5),
    rot=45,
    color="skyblue",
    alpha=0.7,
)
````

We can do something similar with the Performance & Risk metrics as well with the added benefit that it becomes possible to compare the results to the benchmark again. This shows whether you might have a higher or lower risk profile compared to the benchmark. A good way to demonstrate this is through the calculation of the Beta and the Value at Risk.

````python
beta = toolkit.performance.get_beta()

# The benchmark has a beta of 1 by definition
beta["Benchmark"] = 1

beta[["Benchmark", "Portfolio"]].plot(
    title="Beta", figsize=(15, 5), rot=0, color=["black", "green"], grid=True
)

value_at_risk = toolkit.risk.get_value_at_risk()

value_at_risk[["Benchmark", "Portfolio"]].plot(
    title="Value at Risk",
    kind="bar",
    figsize=(15, 5),
    rot=45,
    color=["skyblue", "salmon"],
    alpha=0.7,
    grid=True,
)
````

Furthermore, this can be extended in the Technicals module as well where you can calculate the Moving Averages, RSI, MACD and more for the assets in your portfolio. This can be useful to see whether the assets are overbought or oversold and whether the assets are in a bullish or bearish trend.

````python
from matplotlib import pyplot as plt

rsi = toolkit.technicals.get_relative_strength_index()

rsi.plot(
    subplots=True,
    figsize=(15, 10),
    title="Relative Strength Index",
    alpha=0.7,
    layout=(7, 4),
    legend=False,
)

for ax, col in zip(plt.gcf().axes, rsi.columns):
    ax.axhline(y=70, color="black", linestyle="-")
    ax.axhline(y=30, color="black", linestyle="-")
    ax.set_title(col)
    ax.tick_params(axis="x", rotation=45)
    ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
````
